```python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_csv("health_dataset.csv")

glucose_diabetic = df[df['diabetes'] == True]['glucose_level']
glucose_nondiabetic = df[df['diabetes'] == False]['glucose_level']
t_stat, p_value = ttest_ind(glucose_diabetic, glucose_nondiabetic, equal_var=False)

print("T-value:", t_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject H0: Significant difference in glucose levels.")
else:
    print("Fail to reject H0: No significant difference in glucose levels.")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import t

df = pd.read_csv("health_dataset.csv")
glucose_diabetic = df[df['diabetes'] == True]['glucose_level'].dropna()
glucose_nondiabetic = df[df['diabetes'] == False]['glucose_level'].dropna()

mean1 = glucose_diabetic.mean()
mean2 = glucose_nondiabetic.mean()
std1 = glucose_diabetic.std(ddof=1)
std2 = glucose_nondiabetic.std(ddof=1)

n1 = len(glucose_diabetic)
n2 = len(glucose_nondiabetic)
mean_diff = mean1 - mean2
se = np.sqrt((std1**2 / n1) + (std2**2 / n2))

df_welch = (std1**2 / n1 + std2**2 / n2)**2 / (
    (std1**2 / n1)**2 / (n1 - 1) + (std2**2 / n2)**2 / (n2 - 1)
)
alpha = 0.05
t_critical = t.ppf(1 - alpha/2, df_welch)

lower_ci = mean_diff - t_critical * se
upper_ci = mean_diff + t_critical * se

print(f"95% Confidence Interval: ({lower_ci:.2f}, {upper_ci:.2f})")
```

```python
from scipy.stats import ttest_ind, t

glucose_diabetic = df[df['diabetes'] == True]['glucose_level'].dropna()
glucose_nondiabetic = df[df['diabetes'] == False]['glucose_level'].dropna()

t_stat, p_value = ttest_ind(
    glucose_diabetic,
    glucose_nondiabetic,
    equal_var=False)

n1, n2 = len(glucose_diabetic), len(glucose_nondiabetic)
s1 = glucose_diabetic.std(ddof=1)
s2 = glucose_nondiabetic.std(ddof=1)

df_welch = (s1**2/n1 + s2**2/n2)**2 / (
    (s1**2/n1)**2/(n1-1) + (s2**2/n2)**2/(n2-1))
alpha = 0.05
critical_value = t.ppf(1 - alpha/2, df_welch)

print("Critical Value:", critical_value)
print("P-value:", p_value)
```

```python
import numpy as np
from scipy.stats import ttest_ind, norm

group1 = df[df['diabetes'] == True]['glucose_level'].dropna()
group2 = df[df['diabetes'] == False]['glucose_level'].dropna()

n1, n2 = len(group1), len(group2)
mean1, mean2 = group1.mean(), group2.mean()
std1, std2 = group1.std(ddof=1), group2.std(ddof=1)

alpha = 0.05
se = np.sqrt((std1**2 / n1) + (std2**2 / n2))
z_stat = (mean1 - mean2) / se
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Test Used: Z-test")
print("Z-statistic:", z_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Result: Reject H0")
else:
    print("Result: Fail to Reject H0")
```

```python
import pandas as pd
from scipy.stats import f_oneway

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
bins = [0, 30, 45, 60, 120]
labels = ['<30', '30-45', '45-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

group_1 = df[df['age_group'] == '<30']['diabetes']
group_2 = df[df['age_group'] == '30-45']['diabetes']
group_3 = df[df['age_group'] == '45-60']['diabetes']
group_4 = df[df['age_group'] == '60+']['diabetes']

f_stat, p_value = f_oneway(group_1, group_2, group_3, group_4)
alpha = 0.05
print("F-value:", f_stat)

if p_value < alpha:
    print("Decision: Reject H0")
    print("Conclusion: Disease rates differ significantly across age groups.")
else:
    print("Decision: Fail to Reject H0")
    print("Conclusion: No significant difference in disease rates across age groups.")
```
