```python
import pandas as pd

df=pd.read_csv('Smart_Outcome_Predictor_Dataset_5200.csv')
df.head(5)
```

```python
df.isnull().sum()
```

Fill missing values

```python
df["time_spent_hours"] = df["time_spent_hours"].fillna(df["time_spent_hours"].median())
df["avg_quiz_score"] = df["avg_quiz_score"].fillna(df["avg_quiz_score"].median())
df["attendance_rate"] = df["attendance_rate"].fillna(df["attendance_rate"].median())

print(df.isnull().sum())
```

* Part B: Data understanding & preparation

Identify features and target variables for both tasks.

* Features - student_id,age,week_of_year,sessions,time_spent_hours,videos_watched,quiz_attempts,assignments_submitted,forum_posts,avg_quiz_score,attendance_rate,
country_region,device_type,education_background,course_level,course_category,course_start_date

* Target - completion_status & final_score

```python
X = df.drop(columns=["completion_status", "final_score"])
y_classification = df["completion_status"]
y_regression = df["final_score"]

print('Features Shape:', X.shape)
print('Classification Target Shape:', y_classification.shape)
print('Regression Target Shape:', y_regression.shape)
```

Perform a train-test split for classification and regression problems.

```python
from sklearn.model_selection import train_test_split

X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X,y_classification,test_size=0.2,random_state=42,stratify=y_classification)

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X,y_regression,test_size=0.2,random_state=42)

print('Classification Training Shape:', X_train_cls.shape)
print('Classification Testing Shape:', X_test_cls.shape)
print('Regression Training Shape:', X_train_reg.shape)
print('Regression Testing Shape:', X_test_reg.shape)
```

Apply basic preprocessing (scaling/encoding if required).

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler

numerical_cols = ['age','week_of_year','sessions','time_spent_hours','videos_watched','quiz_attempts','assignments_submitted','forum_posts','avg_quiz_score','attendance_rate']
categorical_cols = ['country_region','device_type','education_background','course_level','course_category']

scaler = StandardScaler()

X_train_num_cls = pd.DataFrame(
    scaler.fit_transform(X_train_cls[numerical_cols]),
    columns=numerical_cols,
    index=X_train_cls.index
)

X_test_num_cls = pd.DataFrame(
    scaler.transform(X_test_cls[numerical_cols]),
    columns=numerical_cols,
    index=X_test_cls.index
)

encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

X_train_cat_cls = pd.DataFrame(
    encoder.fit_transform(X_train_cls[categorical_cols]),
    columns=encoder.get_feature_names_out(categorical_cols),
    index=X_train_cls.index
)

X_test_cat_cls = pd.DataFrame(
    encoder.transform(X_test_cls[categorical_cols]),
    columns=encoder.get_feature_names_out(categorical_cols),
    index=X_test_cls.index
)

X_train_cls = pd.concat([X_train_num_cls, X_train_cat_cls], axis=1)
X_test_cls = pd.concat([X_test_num_cls, X_test_cat_cls], axis=1)

scaler = StandardScaler()

X_train_num_reg = pd.DataFrame(
    scaler.fit_transform(X_train_reg[numerical_cols]),
    columns=numerical_cols,
    index=X_train_reg.index
)

X_test_num_reg = pd.DataFrame(
    scaler.transform(X_test_reg[numerical_cols]),
    columns=numerical_cols,
    index=X_test_reg.index
)

encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

X_train_cat_reg = pd.DataFrame(
    encoder.fit_transform(X_train_reg[categorical_cols]),
    columns=encoder.get_feature_names_out(categorical_cols),
    index=X_train_reg.index
)

X_test_cat_reg = pd.DataFrame(
    encoder.transform(X_test_reg[categorical_cols]),
    columns=encoder.get_feature_names_out(categorical_cols),
    index=X_test_reg.index
)

X_train_reg = pd.concat([X_train_num_reg, X_train_cat_reg], axis=1)
X_test_reg = pd.concat([X_test_num_reg, X_test_cat_reg], axis=1)
X_train_cls.head()
```

* Part C: Bagging (Bootstrap Aggregating): 

Implement Bagging Classifier for course completion prediction.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, classification_report

dt_classifier = DecisionTreeClassifier(random_state=42)
bagging_classifier = BaggingClassifier(
    estimator=dt_classifier,
    n_estimators=100,
    random_state=42
)

bagging_classifier.fit(X_train_cls, y_train_cls)
y_pred_bag_cls = bagging_classifier.predict(X_test_cls)

print('Bagging Classifier Accuracy:', accuracy_score(y_test_cls, y_pred_bag_cls))
print('\nClassification Report:')
print(classification_report(y_test_cls, y_pred_bag_cls))
```

Implement Bagging Regressor for final score prediction.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

dt_regressor = DecisionTreeRegressor(random_state=42)

bagging_regressor = BaggingRegressor(
    estimator=dt_regressor,
    n_estimators=100,
    random_state=42
)

bagging_regressor.fit(X_train_reg, y_train_reg)
y_pred_bag_reg = bagging_regressor.predict(X_test_reg)
print('MAE :', mean_absolute_error(y_test_reg, y_pred_bag_reg))
print('MSE :', mean_squared_error(y_test_reg, y_pred_bag_reg))
print('RMSE:', mean_squared_error(y_test_reg, y_pred_bag_reg) ** 0.5)
print('R² Score:', r2_score(y_test_reg, y_pred_bag_reg))
```

Compare bagging results with a single base model.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

single_classifier = DecisionTreeClassifier(random_state=42)
single_classifier.fit(X_train_cls, y_train_cls)
y_pred_single_cls = single_classifier.predict(X_test_cls)

single_accuracy = accuracy_score(y_test_cls, y_pred_single_cls)
bagging_accuracy = accuracy_score(y_test_cls, y_pred_bag_cls)

from sklearn.metrics import r2_score

single_regressor = DecisionTreeRegressor(random_state=42)
single_regressor.fit(X_train_reg, y_train_reg)
y_pred_single_reg = single_regressor.predict(X_test_reg)

single_r2 = r2_score(y_test_reg, y_pred_single_reg)
bagging_r2 = r2_score(y_test_reg, y_pred_bag_reg)
```

```python
comparison = pd.DataFrame({
    "Model": ['Decision Tree Classifier','Bagging Classifier','Decision Tree Regressor','Bagging Regressor'],
    "Metric": ['Accuracy','Accuracy','R² Score','R² Score'],
    "Score": [single_accuracy,bagging_accuracy,single_r2,bagging_r2]
})

comparison
```

Part E: Voting & Stacking Ensembles

Implement Voting Classifier using multiple base classifiers.

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

dt = DecisionTreeClassifier(random_state=42)
lr = LogisticRegression(max_iter=1000, random_state=42)
knn = KNeighborsClassifier()

hard_voting = VotingClassifier(
    estimators=[
        ('dt', dt),
        ('lr', lr),
        ('knn', knn)
    ],
    voting='hard'
)

hard_voting.fit(X_train_cls, y_train_cls)
y_pred_hard = hard_voting.predict(X_test_cls)
print("Hard Voting Accuracy:", accuracy_score(y_test_cls, y_pred_hard))
```

Compare Hard Voting vs Soft Voting.

```python
soft_voting = VotingClassifier(
    estimators=[
        ('dt', dt),
        ('lr', lr),
        ('knn', knn)
    ],
    voting='soft'
)

soft_voting.fit(X_train_cls, y_train_cls)
y_pred_soft = soft_voting.predict(X_test_cls)

print('Hard Voting Accuracy :', accuracy_score(y_test_cls, y_pred_hard))
print('Soft Voting Accuracy :', accuracy_score(y_test_cls, y_pred_soft))
```

Implement Stacking Classifier with a meta-learner.

```python
from sklearn.ensemble import StackingClassifier

stack_classifier = StackingClassifier(
    estimators=[
        ('dt', DecisionTreeClassifier(random_state=42)),
        ('knn', KNeighborsClassifier()),
        ('lr', LogisticRegression(max_iter=1000))
    ],
    final_estimator=LogisticRegression()
)

stack_classifier.fit(X_train_cls, y_train_cls)
y_pred_stack_cls = stack_classifier.predict(X_test_cls)
print("Stacking Accuracy:",accuracy_score(y_test_cls, y_pred_stack_cls))
```

Implement Stacking Regressor for score prediction.

```python
from sklearn.ensemble import StackingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

stack_regressor = StackingRegressor(
    estimators=[
        ('dt', DecisionTreeRegressor(random_state=42)),
        ('knn', KNeighborsRegressor()),
        ('lr', LinearRegression())
    ],
    final_estimator=LinearRegression()
)

stack_regressor.fit(X_train_reg, y_train_reg)
y_pred_stack_reg = stack_regressor.predict(X_test_reg)
print("MAE :", mean_absolute_error(y_test_reg, y_pred_stack_reg))
print("RMSE:", mean_squared_error(y_test_reg, y_pred_stack_reg)**0.5)
print("R² Score:", r2_score(y_test_reg, y_pred_stack_reg))
```

* Part G: Final Analysis & Reporting

-1. Impact of Bagging vs Boosting

Bagging (Bootstrap Aggregating):

-Bagging trains multiple models independently using different bootstrap samples of the training data. 

-The final prediction is obtained by averaging (regression) or majority voting (classification). 

-It mainly reduces model variance and helps prevent overfitting. 

-Bagging is easy to parallelize because all base models are trained independently. 

-Example: Bagging Classifier and Bagging Regressor.

Boosting:

-Boosting trains models sequentially, where each new model focuses on correcting the errors made by previous models. 

-It reduces both bias and variance, often achieving higher predictive accuracy. 

-Boosting generally provides better performance than bagging but requires more training time. 

-Examples: AdaBoost, Gradient Boosting, LightGBM, and XGBoost.

-2. Comparison of Tree-Based Boosting Algorithms

- AdaBoost:

-Uses weak learners sequentially. 

-Increases the weight of misclassified samples. 

-Performs well on relatively simple datasets.
Sensitive to noisy data and outliers.

- Gradient Boosting:

-Builds trees sequentially by minimizing prediction errors. 

-Produces higher accuracy than AdaBoost. 

-Requires careful tuning of hyperparameters. 

- LightGBM:

-Uses histogram-based tree learning. 

-Faster training and lower memory consumption. 

-Efficient for large datasets with many features. 

-Handles categorical features effectively.

- XGBoost:

-Includes regularization to reduce overfitting. 

-Handles missing values automatically. 

-Highly optimized for speed and accuracy.

-Generally achieves the best predictive performance among boosting algorithms.

-3. Advantages of Voting and Stacking 

- Voting Ensemble: 

-Combines predictions from multiple classifiers. 

-Reduces individual model errors. 

-Improves prediction stability. 

-Soft Voting generally performs better than Hard Voting because it considers prediction probabilities.

- Stacking Ensemble: 

-Combines predictions from several base models using a meta-learner. 

-Learns how to best combine different models. 

-Usually provides higher accuracy than individual models. 

-Reduces the weaknesses of single algorithms.

-4. Final Model Recommendation for Deployment

For Classification: In most cases, XGBoost Classifier or Stacking Classifier provides the best overall performance.

Recommendation: highest accuracy obtained Stacking.

For Regression: In most cases, XGBoost Regressor or Stacking Regressor achieves the best predictive performance. 

Recommendation: highest R² score and lowest error obtained in Stacking
