```python
import pandas as pd
df=pd.read_csv('D:\Maths\student_performance_dataset.csv')
print(df)
```

```python
mean = df["Math_Score"].mean()
median = df["Math_Score"].median()
mode = df["Math_Score"].mode()[0]

range = df["Science_Score"].max() - df["Science_Score"].min()
variance = df["Science_Score"].var()
std = df["Science_Score"].std()

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("\nRange:", range)
print("Variance:", variance)
print("Standard Deviation:", std)
```

```python
p_pass = (df["Pass_Fail"] == 1).mean()

df["Hours_Above_5"] = df["Hours_Studied"] > 5
contingency_table = pd.crosstab(df["Pass_Fail"], df["Hours_Above_5"])
p_pass_given_hours = contingency_table.loc[1, True] / contingency_table[True].sum()

print("Probability of Passing:", p_pass)
print("\nContingency Table:\n", contingency_table)
print("\nP(Pass | Hours_Studied > 5):", p_pass_given_hours)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

plt.figure(figsize=(4,3))
plt.hist(df["Math_Score"], bins=10, density=True)
x = np.linspace(df["Math_Score"].min(), df["Math_Score"].max(), 100)
plt.plot(x, stats.norm.pdf(x, df["Math_Score"].mean(), df["Math_Score"].std()))
plt.title("Math Score Distribution")
plt.show()

print("Skewness:", df["Science_Score"].skew())
print("Kurtosis:", df["Science_Score"].kurt())

plt.figure(figsize=(4,3))
stats.probplot(df["English_Score"], plot=plt)
plt.title("Q-Q Plot - English Score")
plt.show()
```

```python
math_vec = df.loc[:4, "Math_Score"].to_numpy()
science_vec = df.loc[:4, "Science_Score"].to_numpy()
dot_product = math_vec @ science_vec

norm1 = np.linalg.norm(math_vec, 1)
norm2 = np.linalg.norm(math_vec, 2)

cos_theta = dot_product / (np.linalg.norm(math_vec) * np.linalg.norm(science_vec))
angle_deg = np.degrees(np.arccos(cos_theta))

print("Math_Score Vector:", math_vec)
print("Science_Score Vector:", science_vec)
print("Dot Product:", dot_product)
print("Math Vector L1 Norm:", norm1)
print("Math Vector L2 Norm:", norm2)
print("Angle Between Vectors (degrees):", angle_deg)
```
