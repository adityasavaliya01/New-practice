```python
import numpy as np
import pandas as pd
from numpy.linalg import norm, det, inv, eig
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from scipy.linalg import lu, svd
from sklearn.preprocessing import StandardScaler

df=pd.read_csv('D:\Maths\student_scores_linear_algebra.csv',index_col=0)
df
X = df.values
print(X)
```

```python
v1 = X[0]
v2 = X[1]

norm_1 = norm(v1, 1)
norm_2 = norm(v1, 2)

norm_1, norm_2
```

```python
dot_product = np.dot(v1, v2)
angle = np.arccos(dot_product / (norm(v1)*norm(v2))) * (180/np.pi)
dot_product, angle
```

```python
cross_product = np.cross(v1[:3], v2[:3])
cross_product
```

```python
projection = (np.dot(v1, v2) / np.dot(v2, v2)) * v2
projection
```

```python
A = X[:5]
B = X[5:10]

matrix_addition = A + B
matrix_multiplication = np.dot(A, B.T)
matrix_addition, matrix_multiplication
```

```python
square_matrix = X[:5, :5]

transpose = square_matrix.T
determinant = det(square_matrix)
inverse_matrix = inv(square_matrix)
transpose, determinant, inverse_matrix
```

```python
cov_matrix = np.cov(X.T)
eigenvalues, eigenvectors = eig(cov_matrix)

eigenvalues, eigenvectors
```

```python
P, L, U = lu(square_matrix)
P, L, U
```

```python
U_svd, S_svd, Vt_svd = svd(X)
U_svd, S_svd, Vt_svd
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
labels = (df.mean(axis=1) >= df.mean(axis=1).mean()).astype(int)

lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(X_scaled, labels)
X_lda
```
